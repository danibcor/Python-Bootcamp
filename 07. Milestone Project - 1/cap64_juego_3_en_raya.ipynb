```python
from IPython.display import clear_output # Me permite limpiar la consola

num_fila = 3
MAX_NUM = num_fila * num_fila

def fila(f):
    
    jugador_ganador = "Nadie"
    
    if f[0] == f[1] == f[2]:
        if f[0] == 'X':
            jugador_ganador = "Jugador 1"
        elif f[0] == 'O':
            jugador_ganador = "Jugador 2"
        
    return jugador_ganador

def col(f1,f2,f3):
    
    jugador_ganador = "Nadie"
    
    for i in range(0,len(f1)):
        if f1[i] == f2[i] == f3[i]:
            if f1[i] == 'X':
                jugador_ganador = "Jugador 1"
            elif f1[i] == 'O':
                jugador_ganador = "Jugador 2"
            break
    
    return jugador_ganador
    
def diagonal(f1,f2,f3):
    
    jugador_ganador = "Nadie"
    
    if f1[0] == f2[1] == f3[2]:
        if f1[0] == 'X':
            jugador_ganador = "Jugador 1"
        elif f1[0] == 'O':
            jugador_ganador = "Jugador 2"
    elif f1[2] == f2[1] == f3[0]:
        if f1[2] == 'X':
            jugador_ganador = "Jugador 1"
        elif f1[2] == 'O':
            jugador_ganador = "Jugador 2"
            
    return jugador_ganador

def comprobar_ganador(f1,f2,f3):
    
    jugador_ganador = "Nadie"
    
    if fila(f1) != "Nadie":
        jugador_ganador = fila(f1)
    elif fila(f2) != "Nadie":
        jugador_ganador = fila(f2)
    elif fila(f3) != "Nadie":
        jugador_ganador = fila(f3)
    elif col(f1,f2,f3) != "Nadie":
        jugador_ganador = col(f1,f2,f3)
    elif diagonal(f1,f2,f3) != "Nadie":
        jugador_ganador = diagonal(f1,f2,f3)
        
    return jugador_ganador
    
def comprobar_tablero(f1,f2,f3,fil,col):
    
    if fil == 0:
        if f1[col] == ' ':
            return True
    elif fil == 1:
        if f2[col] == ' ':
            return True
    elif fil == 2:
        if f3[col] == ' ':
            return True
    
    return False

def marcar_tablero(f1,f2,f3,fil,col,p):
    
    if p == True:
        if fil == 0:
            f1[col] = 'X'
        elif fil == 1:
            f2[col] = 'X'
        elif fil == 2:
            f3[col] = 'X'
        p = False
    else:
        if fil == 0:
            f1[col] = 'O'
        elif fil == 1:
            f2[col] = 'O'
        elif fil == 2:
            f3[col] = 'O'
        p = True
        
    return f1,f2,f3,p

def pintar_fila(f):
    
    print("   |   |   ")
    for i in range(0,num_fila):
        if i < num_fila - 1:
            print(f" {f[i]} |",end='')
        else:
            print(f" {f[i]} ")
    print("   |   |   ")
            
def dibujar_tablero(f1,f2,f3):
    
    clear_output() # Me permite limpiar todo lo que hay antes del tablero
    print("\t-- TABLERO --\n")
    pintar_fila(f1)
    print("-----------")
    pintar_fila(f2)
    print("-----------")
    pintar_fila(f3)
    print("\n")
    
def pedir_valores(fil,col,p):
    
    if p:
        print("Jugador 1 --")
    else:
        print("Jugador 2 --")
            
    fil = input("Fila: ")
    col = input("Columna: ")
    
    while fil.isdigit() == False or col.isdigit() == False:
        print("\n\t-> El valor introducido no es valido")
        fil = input("Fila: ")
        col = input("Columna: ")
        
    print("\n")
    
    return int(fil),int(col)
        
def elegir_lugar(f1,f2,f3,p):
    
    ganador = "Nadie"
    fil = 0
    col = 0
    cont = 0
    
    dibujar_tablero(f1,f2,f3)
    
    print("\t-- ASIGNACION JUGADORES --\n")
    print("Jugador 1 -> X")
    print("Jugador 2 -> O\n")
    
    while ganador == "Nadie" and cont < MAX_NUM:
        
        fil,col = pedir_valores(fil,col,p)
        
        while (fil < 0 or fil >= num_fila or col < 0 or col >= num_fila) or comprobar_tablero(f1,f2,f3,fil,col) == False:
            
            print("Sitio no disponible")
            fil,col = pedir_valores(fil,col,p)
        
        f1,f2,f3,p = marcar_tablero(f1,f2,f3,fil,col,p)
        dibujar_tablero(f1,f2,f3)
        ganador = comprobar_ganador(f1,f2,f3)
        cont += 1
    
    if cont == MAX_NUM:
        print("No ha ganado nadie")
    else:
        print(f"Ha ganado el {ganador}")

f1 = [' ',' ',' ']
f2 = [' ',' ',' ']
f3 = [' ',' ',' ']
play = True # True jugador 1, False jugador 2

elegir_lugar(f1,f2,f3,play)
```
