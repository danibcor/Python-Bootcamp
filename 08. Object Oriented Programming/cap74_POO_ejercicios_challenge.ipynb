```python
# Para este desafío, cree una clase de cuenta bancaria que tenga dos atributos:
# propietario
# balance

# y dos métodos:
# depósito
# retirarse

# Como requisito adicional, los retiros no pueden exceder el saldo disponible.

# Instale su clase, haga varios depósitos y retiros, y pruebe para asegurarse de que la cuenta no puede 
# estar sobregirada.
```

```python
class Account:
    
    def __init__(self,user,money = 0):
        
        self.user = user
        self.money = money
        
    def __str__(self):
        
        return f"Nombre de la cuenta: {self.user} \nDinero en la cuenta: {self.money}"
        
    def ingresar_dinero(self):
        
        dinero_st = input("\nDinero a ingresar: ")
        
        while dinero_st.isdigit() == False or int(dinero_st) < 0:
            
            dinero_st = input("Dinero a ingresar: ")
        
        self.money += float(dinero_st)
           
    def retirar_dinero(self):
        
        dinero_st = input("\nDinero a retirar: ")
        
        while dinero_st.isdigit() == False or int(dinero_st) < 0 or int(dinero_st) > self.money:
            
            dinero_st = input("Dinero a ingresar: ")
            
        self.money -= float(dinero_st)
            
def pedir_datos():
    
    nombre = input("\nNombre del usuario: ")
    dinero_st = input("Dinero a ingresar: ")
    dinero = 0
        
    while dinero_st.isdigit() == False or int(dinero_st) < 0:
        
        dinero_st = input("Dinero a ingresar: ")
    
    dinero = float(dinero_st) 
        
    return nombre,dinero

def ingresar(mi_cuenta):
    
    opcion = input("\nQuiere ingresar dinero? (S: si , N: no): ").upper()
    
    while opcion != 'S' and opcion != 'N':
        
        print("Opcion introducida no valida")
        opcion = input("Quiere ingresar dinero? (S: si , N: no): ").upper()
        
    while opcion == 'S':
        
        mi_cuenta.ingresar_dinero()
        print(f"\n{mi_cuenta}")
        opcion = input("Quiere ingresar mas dinero? (S: si , N: no): ").upper()
        
def retirar(mi_cuenta):
    
    opcion = input("\nQuiere retirar dinero? (S: si , N: no): ").upper()
    
    while opcion != 'S' and opcion != 'N':
        
        print("Opcion introducida no valida")
        opcion = input("Quiere retirar dinero? (S: si , N: no): ").upper()
        
    while opcion == 'S':
        
        mi_cuenta.retirar_dinero()
        print(f"\n{mi_cuenta}")
        opcion = input("Quiere retirar mas dinero? (S: si , N: no): ").upper()

nombre,dinero = pedir_datos()
mi_cuenta = Account(nombre,dinero)
ingresar(mi_cuenta)
retirar(mi_cuenta)
```
