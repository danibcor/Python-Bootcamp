```python
MAX_PERSONAS = 10
from IPython.display import clear_output # Me permite limpiar la consola

class Coche():
    
    # Atributo Globales de la clase, es decir que cada instancia de la clase comparte estas mismas caracteristicas
    ruedas = 4
    espejos = 2
    puertas = True
        
    def __init__(self,marca,modelo,estado,equipado,contador_acceso,lista_personas):
        
        self.marca = marca
        self.modelo = modelo
        self.estado = estado
        self.equipado = equipado
        self.acceso = contador_acceso
        self.lista_personas= lista_personas
        
    def pedir_info(self):
        
        self.marca = input(f"Marca: ")
        self.modelo = input(f"Modelo: ")
        self.estado = input(f"Estado: ")
        car = input(f"Equipado (S: si , N:no): ").upper()
        
        while car != 'S' and car != 'N':
            print("Caracter no valido")
            car = input(f"Equipado (S: si , N:no): ").upper()
            
        if car == 'S':
            self.equipado = True
        else:
            self.esquipado = False
            
    def mostrar_info(self):
        
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Estado: {self.estado}")
        print(f"Equipado: ",end = '')
        if self.equipado:
            print("Kit completo")
        else:
            print("Sin mejoras")
            
    def personas_acceso(self):
        
        for i in range(0,self.acceso):
            
            print(f"{self.lista_personas[i]} tiene acceso a vehiculo")
            
    def maximo(self):
        
        return self.acceso == MAX_PERSONAS
            
    def comprobar_acceso(self,num):
        
        return self.acceso + num >= 0 and self.acceso + num <= MAX_PERSONAS
            
    def acceso_al_coche(self):
        
        lista_personas = []
        
        if self.maximo() == False:
            
            num_personas = int(input("Numero de personas para agregar acceso: "))

            while self.comprobar_acceso(num_personas) == False or num_personas < 0 or num_personas > MAX_PERSONAS:

                    if num_personas < 0:
                        print("El valor no puede ser negativo")

                    elif num_personas > MAX_PERSONAS or self.comprobar_acceso(num_personas) == False:
                        print(f"El valor no puede superar el maximo de {MAX_PERSONAS - self.acceso}")

                    num_personas = int(input("Numero de personas para agregar acceso: "))
            
            self.acceso += num_personas

            for i in range(0,num_personas):

                self.lista_personas.append(input("Nombre persona: "))
        else:
            
            print("Se ha completado el aforo de personas con acceso")
            
    def encontrar_nombre(self,nombre):
        
        j = 0
        
        while j < self.acceso and self.lista_personas[j] != nombre:
            
            j += 1
            
        return j
            
    def retirar_acceso_vehiculo(self):
        
        name = input("Nombre de la persona para quitar acceso: ")
        
        pos = self.encontrar_nombre(name)
        
        if pos < self.acceso: # Es porque encontre el nombre de la lista
            
            self.lista_personas[pos] = self.lista_personas[self.acceso - 1]
            self.lista_personas[self.acceso - 1] = ""
            self.acceso = self.acceso - 1
            
        else:
            
            print("El nombre introducido no se encuentra en la lista de personas con acceso")
        

def algoritmo(mi_coche):
    
    mi_coche.pedir_info()
    clear_output() # Me permite limpiar todo lo que hay antes de mostrar la info
    mi_coche.mostrar_info()
    print("\n")
    
    while mi_coche.acceso != MAX_PERSONAS:
        mi_coche.acceso_al_coche()
        print("\n")
    
    print("\n")
    mi_coche.personas_acceso()
    print("\n")
    mi_coche.retirar_acceso_vehiculo()
    print("\n")
    mi_coche.personas_acceso()
    
def inicializar():
    
    marca = ""
    modelo = ""
    estado = ""
    equipado = False
    acceso = 0
    lista_personas = []
    
    return marca,modelo,estado,equipado,acceso,lista_personas
        
marca,modelo,estado,equipado,acceso,lista_personas = inicializar()
mi_coche = Coche(marca,modelo,estado,equipado,acceso,lista_personas)

algoritmo(mi_coche)
```
